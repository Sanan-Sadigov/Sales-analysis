```python
sales_df['Product'].nunique()
sales_df.drop_duplicates(inplace=True)
sales_df
```

```python
# Monthly Sales Summary by Product: Total Quantity Ordered
summary=sales_df.groupby(['Product','Month'])['Quantity Ordered'].sum().reset_index()
summary
```

1.	Monthly Sales Analysis:
â€¢	Analyze the total sales for each month to identify seasonal trends.


```python
sales_df['Order Date']=pd.to_datetime(sales_df['Order Date'])
sales_df['Year'] = sales_df['Order Date'].dt.year
sales_df['Day'] = sales_df['Order Date'].dt.day
# converting month to int
sales_df['Month'] = sales_df['Month'].astype(int)

# Creating the 'Date' column
sales_df['Date'] = pd.to_datetime(dict(year=sales_df['Year'], month=sales_df['Month'], day=sales_df['Day']))

# Defining a function to determine the season
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Applying the function to the 'Month' column
sales_df['Season'] = sales_df['Month'].apply(get_season)

Seasonal_sales=sales_df[['Product','Sales','Season']]
Seasonal_sales
```

2.Product Performance Analysis:
â€¢ Determine the top-selling products by analyzing quantity ordered and sales amount.
â€¢ Identify the most profitable products based on sales revenue.


```python
# First way
Top_products=sales_df.groupby('Product')['Quantity Ordered'].max().reset_index()
Most_selled_products=sales_df.groupby('Product')['Sales'].max().reset_index()
Most_selled_products
```

3.	City-wise Sales Analysis:
â€¢	Explore sales data based on different cities to identify regions with the highest sales volume.


```python
Top_cities=sales_df.groupby('City')['Order ID'].count().reset_index(name='Order Count')
Top_cities
```

```python
# Exporting Multiple Sales Data Summaries to Excel with Separate Sheets
with pd.ExcelWriter(r'c:\Users\Senan\Desktop\Sales data new project.xlsx', engine='xlsxwriter') as writer:
    Most_selled_products.to_excel(writer, sheet_name='Most_selled_products', index=False)
    Top_cities.to_excel(writer, sheet_name='Top_cities', index=False)
    Seasonal_sales.to_excel(writer, sheet_name='Seasonal_sales', index=False)
```
